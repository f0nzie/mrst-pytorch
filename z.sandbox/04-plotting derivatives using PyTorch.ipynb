## Tensor with one element and its derivative

## Plotting function and gradient, `y=sin(x)`

```
import torch
import numpy as np
import matplotlib.pyplot as plt  

x = torch.linspace(-np.pi, np.pi, 100, requires_grad=True)
w = torch.ones(x.size()) # tensor of one same size as x

# the function
y = torch.sin(x)

# compute the derivative
y.backward(w) # handing over the parameter x, as y isn't a scalar anymore

# your function
plt.plot(x.detach().numpy(), y.detach().numpy(), label='sin x')

# gradients
plt.plot(x.detach().numpy(), x.grad.detach().numpy(), label='grad')

plt.legend()
```

## Plotting derivative of `log(x)`

```
import torch
import matplotlib.pyplot as plt  

x = torch.linspace(-10, 10, 100, requires_grad=True)
w = torch.ones(x.size())
y = torch.log(x)    # removed the sum to stay with the same dimensions

# compute derivative
y.backward(w) # handing over the parameter `w`, as `x` isn't a scalar anymore

# plot function
plt.plot(x.detach().numpy(), y.detach().numpy(), label='log(x)')
# plot derivative 
plt.plot(x.detach().numpy(), x.grad.detach().numpy(), label='grad')
plt.legend()
```

## Plotting derivatives of `x^3`

```
import torch
import numpy as np
import matplotlib.pyplot as plt  

x = torch.linspace(-10, 10, 100, requires_grad=True)
w = torch.ones(x.size()) # tensor of one same size as x
y = x**3
y.backward(w) # handing over the parameter x, as y isn't a scalar anymore

# your function
plt.plot(x.detach().numpy(), y.detach().numpy(), label='x^3')

# gradients
plt.plot(x.detach().numpy(), x.grad.detach().numpy(), label='grad')

plt.legend()
```
