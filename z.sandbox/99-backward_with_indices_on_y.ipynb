## Example 2

```
import torch
import numpy as np 

x = np.arange(1,3,1)
x = torch.from_numpy(x).reshape(len(x), 1)
x = x.float() # torch.Size([2, 1])
x.requires_grad = True
print(x)
w = torch.randn((2,2), requires_grad = True) # torch.Size([2, 2])
print(w)
y = w @ x
print(y)

# Compute gradients of w1 . x
y[0].backward(retain_graph=True)
print(w.grad)
# tensor([[1., 2.],
#        [0., 0.]])
y[1].backward(retain_graph=True)
print(w.grad)
```

```
# this would be the right way of calculating the derivative
y.backward(x, retain_graph=True)
x.grad
```
